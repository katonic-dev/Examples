## Boston House Price Prediction

A model for predicting the house price using some of the regression techniques based on some of features in the dataset which is called **Boston House Price Prediction**.

- In this Notebook, house prices will be predicted on given explanatory variables that cover many aspects of residential houses. 
- The goal is to create a regression model that is able to accurately estimate the price of the house given the features.

### Algorithms

- Katonic Python SDK for Complete ML Model Life Cycle.
- The Auto ML component in the Katonic SDK can train Machine Learning models with just one or two lines of code.
- All the metrics of Regression will get catalouged using SDK.

#### Imports

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from katonic.log.client import load_model
from katonic.ml.regression import Regressor
```

#### Creating an experiment

```python
exp_name = 'boston_prediction'
```

#### Import the dataset

```python
# loading Boston Housing data from github
df = pd.read_csv('https://raw.githubusercontent.com/selva86/datasets/master/BostonHousing.csv')
```

### Data Prepration

```python
df.head()
```

```python
df.shape
```

```python
# Checking null values for training dataset
df.isnull().sum()
```

 There are no missing values in this dataset

### Exploratory Data Analysis

we will use some visualizations to understand the relationship of the target variable with other features.

```python
sns.set(rc={'figure.figsize': (10, 8)})
sns.distplot(df['medv'], bins=30)
sns.set_theme(style="darkgrid");
```

**The values of MEDV (target variable) are distributed normally with few outliers.**

#### we create a correlation matrix that measures the linear relationships between the variables.

```python
correlation_matrix = df.corr().round(2)
# annot = True to print the values inside the square
sns.heatmap(data=correlation_matrix, annot=True);
```

- To fit a regression model, we select those features which have a high correlation with our target variable MEDV. 
- By looking at the correlation matrix we can see that `RM` has a strong positive correlation with `MEDV (0.7)` where as `LSTAT` has a high negative correlation with `MEDV(-0.74)`.
- The features `RAD`, `TAX` have a correlation of 0.91. These feature pairs are strongly correlated to each other.Same with `DIS` and `AGE` which have a correlation of -0.75.

#### Using a scatter plot lets see how `RM` and `LSTAT` features vary with MEDV (target variable).

```python
plt.figure(figsize=(20, 5))

# features = df['lstat'], df['rm']
features = ['lstat', 'rm']
target = df['medv']

for i, col in enumerate(features):
    plt.subplot(1, len(features) , i+1)
    x = df[col]
    y = target
    plt.scatter(x, y, marker='o')
    plt.title(col)
    plt.xlabel(col)
    plt.ylabel('MEDV')
```

- The prices increase as the value of `RM` increases linearly. There are few outliers and the data seems to be capped at 50.
- The prices tend to decrease with an increase in `LSTAT`. Though it doesnt look to be following exactly a linear line.

## Train/Test split

```python
x = df.drop(columns=['medv'], axis=1)
y = df['medv']
```

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=.20,random_state=98)
```

### Regression Model

```python
features = list(X_train.columns)
reg = Regressor(X_train,X_test,y_train,y_test, exp_name, source_name='boston-house-prediction.ipynb', features=features)
```

#### Get registered experiment details

```python
exp_id = reg.id
print("experiment name : ", reg.name)
print("experiment location : ", reg.location)
print("experiment id : ", reg.id)
print("experiment status : ", reg.stage)
```

```python
run_list = reg.search_runs(exp_id)['run_id'].tolist()
if run_list:
    reg.delete_run_by_id(run_list)
```

## Runs of the experiment

```python
df_runs = reg.search_runs(exp_id)
print("Number of runs done : ", len(df_runs))
df_runs
```

#### selecting top Runs on the basis of the metrics R2

```python
top_runs = df_runs.sort_values(['metrics.R2'],ascending=False)
top_runs.head()
```

Thus, we can see all the models and their metrics being logged and can be accessed according to the requirements.
here we have considered `R-squared` as the decision metrics according to which `GradientBoosting Regression` is the best model for the usecase.

## Selecting Best Model

```python
artifacts = top_runs.iloc[0]["artifact_uri"]
run_id = top_runs.iloc[0]["run_id"]
model_name = top_runs.iloc[0]["run_name"] 


print('Best model_artifacts :',artifacts)
print("=" * 100)
print('Best model run_id :',run_id)
print("=" * 100)
print('Best model :',model_name)
print("=" * 100)
print("Best model experiment id :",exp_id)
```

## Registering Best Model

```python
result = reg.register_model(
    run_id=run_id,
    model_name = model_name,\
)
```

```python
reg.change_stage(
    model_name=model_name,
    ver_list = [1],
    stage='Production'
)
```

## Fetching Model

```python
location = f"{artifacts}/{model_name}"
model = load_model(location)
```

## Predict

```python
y_pred = model.predict(X_test)
```

```python
# Prepare variable as DataFrame in pandas
df = pd.DataFrame(X_test)

# Add the target variable to df
df["y_pred"] = y_pred
```
